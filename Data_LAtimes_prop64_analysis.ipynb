```python
#props.where(props['prop_name']=='PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.').dropna()
prop = props[props.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']
prop.info()
```

```python
contribs.info()
```

```python
merged = pd.merge(prop, contribs, on='calaccess_committee_id')
merged.head(3)
```

```python
merged.info()
```

```python
merged['amount'].sum()
```

```python
merged['committee_position'].value_counts()
```

```python
#support  = merged[merged.committee_position=='SUPPORT']
#support.amount.sum()
#support
support = merged.where(merged['committee_position']=='SUPPORT').dropna(subset=['committee_position'])
support['amount'].sum()
```

```python
support.head(3)
```

```python
#oppose = merged[merged.committee_position=='OPPOSE']
#oppose
oppose = merged.where(merged['committee_position'] == 'OPPOSE').dropna(subset=['committee_position'])
oppose['amount'].sum()
```

```python
print 'percentage is:'
support['amount'].sum()/merged['amount'].sum()
```

```python
print 'Max amount obtained by support and oppose:'
print support['amount'].max()
print oppose['amount'].max()
print 'Min amount obtained by support and oppose:'
print support['amount'].min()
print oppose['amount'].min()
```

```python
print 'here is a boxplot:'
plt.figure(figsize=(3, 4))
ax1 = plt.subplot(1,1,1)
ax1.boxplot([support['amount'], oppose['amount']])
ax1.set_title('Support and Oppose boxplots', alpha=1, size=10)
ax1.set_ylabel('Dollars in millions', alpha = 0.8)
ax1.set_xlabel('Group', alpha=0.8)
plt.setp(ax1,yticks=[0,1000000,2000000,3000000,4000000], yticklabels=[0, 1,2,3,4])
plt.setp(ax1, xticks=[1,2], xticklabels=['support','oppose'],)
plt.tight_layout()
#plt.savefig('boxplots.pdf')
#plt.savefig('boxplots.png')
#plt.savefig('boxplots2.pdf', bbox_inches='tight')
```
